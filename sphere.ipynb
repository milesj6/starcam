```
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd



# Make sphere
u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi, 100)

x = 10 * np.outer(np.cos(u), np.sin(v))
y = 10 * np.outer(np.sin(u), np.sin(v))
z = 10 * np.outer(np.ones(np.size(u)), np.cos(v))

#Make points
size = 99
stars = 10

xx = [30 * np.cos(u[randint(0,size)]) * np.sin(v[randint(0,size)]) for i in range(stars)]
yy = [30 * np.sin(u[randint(0,size)]) * np.sin(v[randint(0,size)])  for i in range(stars)]
zz = [30 * np.cos(v[randint(0,size)])  for i in range(stars)]


x_markers = [10,-10,0,0,0,0]
y_markers = [0,0,10,-10,0,0]
z_markers = [0,0,0,0,10, -10]


# Plot the surface
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(x, y, z, color='b',alpha=0.11)
ax.scatter(xx,yy,zz,color="g",s=20,)
ax.scatter(x_markers,y_markers,z_markers,color="r",s=20,)

ax.set_aspect('auto')
plt.show()
```

```
#Sphere data conversion
x_series = pd.Series(x.flatten())
y_series = pd.Series(y.flatten())
z_series = pd.Series(z.flatten())


#Star data conversion
xx_series = pd.Series(xx)
yy_series = pd.Series(yy)
zz_series = pd.Series(zz)

#create dataframe
sphere_data={'x':x_series,'y':y_series,'z':z_series}
sphere_coordinates = pd.DataFrame(data=sphere_data) 

star_data = {'x':xx_series,'y':yy_series,'z':zz_series}
star_coordinates = pd.DataFrame(data=star_data) 
```
